```python
import numpy as np
import math
from math import e
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import networkx as nx


# Crear un grafo vacío
G = nx.Graph()

# Agregar nodos (sitios turísticos)
G.add_nodes_from(['Museo de Antioquia', 'Parque de los Deseos', 'Plaza Botero', 'Pueblito Paisa'])

# Agregar aristas (ciclorrutas)
G.add_edges_from([('Museo de Antioquia', 'Parque de los Deseos'), ('Parque de los Deseos', 'Plaza Botero'), ('Plaza Botero', 'Pueblito Paisa'), ('Pueblito Paisa', 'Pueblito Paisa'), ('Parque de los Deseos', 'Pueblito Paisa'), ('Museo de Antioquia', 'Plaza Botero')])

# Visualizar el grafo
nx.draw(G, with_labels=True)

# Get the adjacency matrix
adjacency_matrix = nx.adjacency_matrix(G)

# Print the adjacency matrix
print(adjacency_matrix.todense())
print(adjacency_matrix)

```
